```python
import pandas as pd
import seaborn as sns
water_levels = pd.read_excel("Mattawa Water Levels.xlsx")
water_levels.tail()
```

```python
# Keeps only the row of means
water_levels_means=water_levels.tail(1)

# removes the row of means and the data for 2022
water_levels_annual=water_levels.head(water_levels.shape[0]-2)
```

```python
for col in water_levels_annual.columns:
    water_levels_annual[col]=water_levels_annual[col].astype(float)
water_levels_annual.dtypes
```

```python
sns.lineplot(data=water_levels_annual, x="Year", y="Annual Means").set_title('Mattawa Annual Mean Water Level (m)')
```

```python
water_level_list=[]
for y,year in enumerate(water_levels_annual):
    yearData=water_levels_annual.iloc[y]
    label_year=str(int(yearData["Year"]))
    for month in ["Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"]:
        water_level_list.append([label_year+" "+month,yearData[month]])
water_level_list
```

```python
import matplotlib.pyplot as plt

levels = [item[1] for item in water_level_list]
dates  = [item[0] for item in water_level_list]
water_level_series = pd.Series(levels, index =dates)

fig, ax = plt.subplots(figsize=(50,8))
sns.lineplot(data=water_level_series,ax=ax).set_title('Mattawa Monthly Mean Water Level (m)')
plt.xticks(rotation=45)
plt.show()
```

```python
water_level_df=pd.DataFrame(water_level_series).reset_index()
water_level_df.columns=["Month","Level (m)"]
ax=sns.regplot(data=water_level_df,x=water_level_df.index,y="Level (m)").set_title('Trend of Monthly Mean Water Level (m) at Mattawa')
plt.xlabel("Months after January 1950")
```

```python
import numpy as np
plt.figure(figsize=(10, 10))
x_month=water_level_df.index
y_level=water_level_df["Level (m)"]
m, b = np.polyfit(x_month, y_level, 1)
plt.plot(x_month, m*x_month + b,'r-',label='Line of Best Fit: Slope={}'.format(round(m,3)))
plt.scatter(x_month,y_level)
plt.ylim([0, 160])
plt.xlabel("month")
plt.ylabel("level (m)")
# plt.title("km run v.s. kg weight loss")
plt.legend()
```

```python
from sklearn.metrics import r2_score
y_pred = m*x_month + b
y_true = y_level
r2_score(y_true, y_pred)
```

AHA! We have a small effect size!

```python
# Adding line where slope = 0 

import numpy as np
plt.figure(figsize=(10, 10))
x_month=water_level_df.index
y_level=water_level_df["Level (m)"]
m, b = np.polyfit(x_month, y_level, 1)
mean_y=[np.mean(y_level)]*len(x_month)
plt.plot(x_month, m*x_month + b,'r-',label='Line of Best Fit: Slope={}'.format(round(m,3)))
plt.plot(x_month, mean_y,'g-',label='Line of Best Fit: Slope=0 (Null Hypothesis)')
plt.scatter(x_month,y_level)
plt.ylim([150, 156])
plt.xlabel("month")
plt.ylabel("level (m)")
# plt.title("km run v.s. kg weight loss")
plt.legend()
```

```python
def sumOfSquares(list1,list2):
    return np.sum(np.square(np.abs(list1-list2)))

# Sum of Squares for Regression (SSR)
SSR = sumOfSquares(y_pred,mean_y)
print("SSR",SSR)
# Total Sums of Squares (TSS)
TSS = sumOfSquares(y_level,mean_y)
print("TSS",TSS)
# Sums of Squares for Error (SSE)
SSE = sumOfSquares(y_level,y_pred)
print("SSE",SSE)

# Degrees of freedom for SSE
dof_sse = len(y_level)-1-1

# Degrees of freedom for SSR
dof_ssr = 1

f_stat = (SSR/dof_ssr)/(SSE/dof_sse) # note, this is numerator over denominator, so later on the top part is the numerator
f_stat
```

```python
# From: https://stackoverflow.com/questions/58497385/calculating-f-statistic-in-sklearn

from sklearn.linear_model import LinearRegression
X_to_fit = np.array(list(x_month))[:,np.newaxis]
y_to_fit = np.array(list(y_true))[:,np.newaxis]
model=LinearRegression().fit(X_to_fit, y_to_fit)

Rsq = model.score(X_to_fit,y_to_fit)
print(Rsq,"Nice, the same R squared!")
# Number of observations
N=len(x_month)
# Number of variables
K=1

fstat = (Rsq/(1-Rsq))*((N-K-1)/K)
print(fstat,"Nice, the same F statistic")
```

Now, what is our maximum p-value? Let's say it is p<.05!

Here "p" tells you the probability the null hypothesis is true but we said it is false

If we know p<0.05 then we are 95% sure we are right

So, ... here we go

```python
from scipy.stats import f
p_value=1-f.cdf(f_stat, dof_ssr,dof_sse)
p_value
```

0.0038 is less than 0.05

NICE!

We conclude that, based on our linear model's fit to the data, we do in fact believe water levels are rising!
