```python
import nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('gutenberg')
nltk.download('inaugural')
nltk.download('brown')
nltk.download('webtext')
```

```python
!cat FUSE/data/dataset_sample.txt|wc -l
```

```python
from nltk.corpus import gutenberg, inaugural, brown, webtext

for fileid in [gutenberg.fileids()]:
  gutenbergSents = [" ".join(s) for s in gutenberg.sents(fileid)]
  print(gutenbergSents[0])
for fileid in [inaugural.fileids()]:
  inauguralSents = [" ".join(s) for s in inaugural.sents(fileid)]
  print(inauguralSents[0])
for fileid in [brown.fileids()]:
  brownSents = [" ".join(s) for s in brown.sents(fileid)]
  print(brownSents[0])
for fileid in [webtext.fileids()]:
  webtextSents = [" ".join(s) for s in webtext.sents(fileid)]
  print(webtextSents[0])
mergedDataset = gutenbergSents + inauguralSents + brownSents + webtextSents

len(mergedDataset)
```

```python
with open("FUSE/data/dataset_sample.txt","w") as f:
  for sent in mergedDataset:
    f.write(sent+"\n")
!cat FUSE/data/dataset_sample.txt|wc -l
```
